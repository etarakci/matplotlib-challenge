## Observations and Insights

1) The final tumor volume of Capomulin and Ramicane are a little over half of the final tumor volume for Infubinol and Ceftamin, indicating that these two drugs may be the best two for decreasing tumor volume.
2) On Capomulin, the biggest decrease in the tumor volume for mouse b128 was from timepoint 25 to 35.
3) There is a moderate positive correlation between mouse weight and the averagetumor volume for mice on Capomulin (r=.659)

## Dependencies and starter code

```python
#Matplotlib hw by Erendiz Tarakci
#last updated 4/4/2020

# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np

# Study data files
mouse_metadata = "data/Mouse_metadata.csv"
study_results = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata)
study_results = pd.read_csv(study_results)

# Combine the data into a single dataset
data = pd.merge(mouse_metadata,study_results,on="Mouse ID")
data = data.rename(columns={"Mouse ID": "Mouse_ID","Drug Regimen":"Drug_Regimen","Weight (g)":"Weight_g","Tumor Volume (mm3)":"Tumor_Volume_mm3","Metastatic Sites":"Metastatic_Sites"})
data.head()
```

## Summary statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
mean = data.groupby("Drug_Regimen")["Tumor_Volume_mm3"].mean()
median = data.groupby("Drug_Regimen")["Tumor_Volume_mm3"].median()
var = data.groupby("Drug_Regimen")["Tumor_Volume_mm3"].var()
st_dev = data.groupby("Drug_Regimen")["Tumor_Volume_mm3"].std()
sem = data.groupby("Drug_Regimen")["Tumor_Volume_mm3"].sem()
count = data.groupby("Drug_Regimen")["Tumor_Volume_mm3"].count()

mean_df = pd.DataFrame(mean).rename(columns={"Tumor_Volume_mm3":"Mean"})
median_df = pd.DataFrame(median).rename(columns={"Tumor_Volume_mm3":"Median"})
var_df = pd.DataFrame(var).rename(columns={"Tumor_Volume_mm3":"Variance"})
st_dev_df = pd.DataFrame(mean).rename(columns={"Tumor_Volume_mm3":"St_Dev"})
sem_df = pd.DataFrame(mean).rename(columns={"Tumor_Volume_mm3":"SEM"})
count_df = pd.DataFrame(count).rename(columns={"Tumor_Volume_mm3":"Count"})

regimen_stats = pd.merge(mean_df,median_df,on="Drug_Regimen")
regimen_stats = pd.merge(regimen_stats,var_df,on="Drug_Regimen")
regimen_stats = pd.merge(regimen_stats,st_dev_df,on="Drug_Regimen")
regimen_stats = pd.merge(regimen_stats,sem_df,on="Drug_Regimen")
regimen_stats_df = pd.merge(regimen_stats,count_df,on="Drug_Regimen")

regimen_stats_df
```

## Quartiles, outliers and boxplots

```python
# Calculate the final tumor volume of each mouse across four of the most promising treatment regimens. Calculate the IQR and quantitatively determine if there are any potential outliers. 

last_timepoint_df = pd.DataFrame(data.groupby("Mouse_ID")["Timepoint"].max())
last_tumor_vol = pd.merge(last_timepoint_df,data,on=['Mouse_ID','Timepoint'], how='left')
last_tumor_vol = last_tumor_vol[["Mouse_ID","Timepoint","Tumor_Volume_mm3","Drug_Regimen"]].copy()

#top 4 treatments = [Capomulin, Ramicane, Infubinol, Ceftamin]
top_four = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']
bottom_six = ['Ketapril','Naftisol','Placebo','Propriva','Stelasyn','Zoniferol']

#Get rid of other treatments
for drug in bottom_six:
    last_tumor_vol = last_tumor_vol[last_tumor_vol.Drug_Regimen != drug]

#Make DF per top 4 treatment (just in case)
for drug in top_four:
    if drug == 'Capomulin':
        cap_tumor_vol = last_tumor_vol[last_tumor_vol.Drug_Regimen == drug].copy()
    elif drug == 'Ramicane':
        ram_tumor_vol = last_tumor_vol[last_tumor_vol.Drug_Regimen == drug].copy()
    elif drug == 'Infubinol':
        inf_tumor_vol = last_tumor_vol[last_tumor_vol.Drug_Regimen == drug].copy()
    elif drug == 'Ceftamin':
        cef_tumor_vol = last_tumor_vol[last_tumor_vol.Drug_Regimen == drug].copy()
        
```

```python
# #Quartile calculations for Tumor Volumes across the top 4 drug treatments
quartiles = last_tumor_vol.Tumor_Volume_mm3.quantile([.25,.5,.75])
lowerq = quartiles[0.25]
upperq = quartiles[0.75]
iqr = upperq-lowerq

print(f"The lower quartile of tumor volumes across 4 treatments is: {lowerq}")
print(f"The upper quartile of tumor volumes across 4 treatments is: {upperq}")
print(f"The interquartile range of tumor volumes across 4 treatments is: {iqr}")
print(f"The the median of tumor volumes across 4 treatments is: {quartiles[0.5]} ")

lower_bound = lowerq - (1.5*iqr)
upper_bound = upperq + (1.5*iqr)
print(f"Values below {lower_bound} could be outliers.")
print(f"Values above {upper_bound} could be outliers.")
        
```

```python
#Quartile Calculatiosn per drug treatment

#Quartile calculations for Capomulin
cap_quartiles = cap_tumor_vol.Tumor_Volume_mm3.quantile([.25,.5,.75])
cap_lowerq = cap_quartiles[0.25]
cap_upperq = cap_quartiles[0.75]
cap_iqr = cap_upperq-cap_lowerq

print("---------Capomulin---------")
print(f"The lower quartile of Capomulin tumor volumes is: {cap_lowerq}")
print(f"The upper quartile of Capomulin tumor volumes is: {cap_upperq}")
print(f"The interquartile range of Capomulin tumor volumes is: {cap_iqr}")
print(f"The the median of Capomulin tumor volumes is: {cap_quartiles[0.5]} ")
cap_lower_bound = cap_lowerq - (1.5*cap_iqr)
cap_upper_bound = cap_upperq + (1.5*cap_iqr)
print(f"Values below {cap_lower_bound} could be outliers.")
print(f"Values above {cap_upper_bound} could be outliers.")
print("\n")

#Quartile calculations for Ramicane
ram_quartiles = ram_tumor_vol.Tumor_Volume_mm3.quantile([.25,.5,.75])
ram_lowerq = ram_quartiles[0.25]
ram_upperq = ram_quartiles[0.75]
ram_iqr = ram_upperq-ram_lowerq

print("---------Ramicane---------")
print(f"The lower quartile of Ramicane tumor volumes is: {ram_lowerq}")
print(f"The upper quartile of Ramicane tumor volumes is: {ram_upperq}")
print(f"The interquartile range of Ramicane tumor volumes is: {ram_iqr}")
print(f"The the median of Ramicane tumor volumes is: {ram_quartiles[0.5]} ")
ram_lower_bound = ram_lowerq - (1.5*ram_iqr)
ram_upper_bound = ram_upperq + (1.5*ram_iqr)
print(f"Values below {ram_lower_bound} could be outliers.")
print(f"Values above {ram_upper_bound} could be outliers.")
print("\n")
        
#Quartile calculations for Infubinol
inf_quartiles = inf_tumor_vol.Tumor_Volume_mm3.quantile([.25,.5,.75])
inf_lowerq = inf_quartiles[0.25]
inf_upperq = inf_quartiles[0.75]
inf_iqr = inf_upperq-inf_lowerq

print("---------Infubinol---------")
print(f"The lower quartile of Infubinol tumor volumes is: {inf_lowerq}")
print(f"The upper quartile of Infubinol tumor volumes is: {inf_upperq}")
print(f"The interquartile range of Infubinol tumor volumes is: {inf_iqr}")
print(f"The the median of Infubinol tumor volumes is: {inf_quartiles[0.5]} ")
inf_lower_bound = inf_lowerq - (1.5*inf_iqr)
inf_upper_bound = inf_upperq + (1.5*inf_iqr)
print(f"Values below {inf_lower_bound} could be outliers.")
print(f"Values above {inf_upper_bound} could be outliers.")
print("\n")

#Quartile calculations for Ceftamin
cef_quartiles = cef_tumor_vol.Tumor_Volume_mm3.quantile([.25,.5,.75])
cef_lowerq = cef_quartiles[0.25]
cef_upperq = cef_quartiles[0.75]
cef_iqr = cef_upperq-cef_lowerq

print("---------Ceftamin---------")
print(f"The lower quartile of Ceftamin tumor volumes is: {cef_lowerq}")
print(f"The upper quartile of Ceftamin tumor volumes is: {cef_upperq}")
print(f"The interquartile range of Ceftamin tumor volumes is: {cef_iqr}")
print(f"The the median of Ceftamin tumor volumes is: {cef_quartiles[0.5]} ")
cef_lower_bound = cef_lowerq - (1.5*cef_iqr)
cef_upper_bound = cef_upperq + (1.5*cef_iqr)
print(f"Values below {cef_lower_bound} could be outliers.")
print(f"Values above {cef_upper_bound} could be outliers.")
```

```python
# Generate a box plot of the final tumor volume of each mouse across four regimens of interest
fig1, ax1 = plt.subplots(figsize=(10,5))
ax1.set_title('Final Tumor Volume Across Top 4 Drug Regimens')
ax1.set_ylabel('Final Tumor Volume (mm3)')
ax1.set_xlabel('Drug Regimen')
vol_data = [cap_tumor_vol.Tumor_Volume_mm3,ram_tumor_vol.Tumor_Volume_mm3,inf_tumor_vol.Tumor_Volume_mm3,cef_tumor_vol.Tumor_Volume_mm3]
green_diamond = dict(markerfacecolor='g', marker='D')
ax1.boxplot(vol_data,labels=top_four,flierprops=green_diamond)
plt.savefig('box plot')
plt.show()
```

```python
# Generate a scatter plot of mouse weight versus average tumor volume for the Capomulin regimen
#avg_tumor_vol = pd.DataFrame(data.groupby("Mouse_ID")["Tumor_Volume_mm3"].mean())
cap_data = data[data.Drug_Regimen == "Capomulin"].copy()
cap_avg = data.groupby(['Mouse_ID']).mean()
plt.scatter(cap_avg.Weight_g,cap_avg.Tumor_Volume_mm3)
plt.xlabel('Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')
plt.title("Mouse Weight vs Avg Tumor Volume (Capomulin)")
plt.savefig('scatterplot')
plt.show()
```

```python
# Calculate the correlation coefficient and linear regression model for mouse weight and average tumor volume for the Capomulin regimen
import scipy.stats as st
from scipy.stats import linregress

#replot scatter plot

x_values = cap_avg.Weight_g #x_values
y_values = cap_avg.Tumor_Volume_mm3 #y_values

(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(20,35),fontsize=15,color="red",annotation_clip=False)

plt.xlabel('Weight (g)')
plt.ylabel('Tumor Volume (mm3)')
plt.title("Mouse Weight vs Avg Tumor Volume (Capomulin)")
print(f"The r-squared is: {rvalue}")
plt.savefig('scatterplot with regression')
plt.show()
```
